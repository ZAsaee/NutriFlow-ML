```python
import boto3
```

```python
s3 = boto3.resource("s3")
bucket_name = "nutriflow-pipeline"
```

```python
try:
    for obj in s3.Bucket(bucket_name).objects.all():
        print(obj.key)
    print("S3 access successful")
except Exception as e:
    print("s3 access failed:", e)
```
